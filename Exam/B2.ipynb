```python
import torch
```

```python
gamma = 0.1
```

```python
f_list = torch.zeros(100, 100, 2)

for i, xi in enumerate(torch.linspace(-1.0, 1.0, steps=100)):
    for j, xj in enumerate(torch.linspace(-1.0, 1.0, steps=100)):

        q1 = torch.exp(-gamma * (xi - xj)**2)
        q2 = -2 * gamma * (xi - xj) * torch.exp(-gamma * (xi - xj)**2)
        q3 = -2 * gamma * (xi - xj) * torch.exp(-gamma * (xi - xj)**2)
        q4 = torch.exp(-gamma * (xi - xj)**2) * (2 * gamma - 4 * gamma**2 * (xi - xj)**2)

        q1 = torch.clamp(q1, min=0.0)
        q2 = torch.clamp(q2, min=0.0)
        q3 = torch.clamp(q3, min=0.0)
        q4 = torch.clamp(q4, min=0.0)

        K = torch.tensor([[q1, q2], [q3, q4]])

        try:
            f = torch.distributions.MultivariateNormal(torch.zeros(2), K).sample()
        except ValueError:
            # PSD error
            K += 0.071* torch.eye(2)
            f = torch.distributions.MultivariateNormal(torch.zeros(2), K).sample()

        f_list[i, j, :] = f
```

```python
print(f_list)
```

```python
X = torch.linspace(-1.0, 1.0, steps=100)
for i in range(2,99):
    gtf = lambda x: -x**2 + 2 * (1)/(1 + torch.exp(-10 * x))
    gtfp = (gtf(X[i+1]) - gtf(X[i-1]))/(X[i+1] - X[i-1])

    print(f"X: {X[i]:.3f}, f: {f_list[i,i,0]:.3f}, gtf: {gtf(X[i]):.3f}, f': {f_list[i,i,1]:.3f}, gtf': {gtfp:.3f}")
```
